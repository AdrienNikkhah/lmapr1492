```python
from pymatgen.ext.matproj import MPRester
from pymatgen.core.operations import SymmOp
from IPython.display import Image
from IPython.display import Math
from IPython.display import Latex
with MPRester("rdmkQ122BgAvNKpjopTO") as m:
    structure = m.get_structure_by_material_id("mp-665")
    structure.to(filename="C_mp-665.cif")
print("1) Axe de rotation d'ordre 3 selon c")
print(" ")
S1 = SymmOp.from_xyz_string("-y,x-y,z")
print("Matrice de rotation associée :")
print(S1.rotation_matrix)
print("Vecteur de translation associé :")
print(S1.translation_vector)
print("Application de l'axe de rotation C3 à l'atome de Se (#2) situé en (x,y,z) = (0.67,0.33,0.23) :")
print(S1.operate((0.67,0.33,0.23)))
Image("C3.jpg")
```

```python
%%latex
Equation matricielle

\begin{equation}
\begin{pmatrix}0 & -1 & 0\\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}\begin{pmatrix}x\\y\\z\end{pmatrix}=\begin{pmatrix}-y\\x-y\\z\end{pmatrix}
\end{equation}

Ce qui donne avec $(x,y,z) = (0.67,0.33,0.23)$ :

\begin{equation}
\begin{pmatrix}0 & -1 & 0\\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}\begin{pmatrix}0.67\\0.33\\0.23\end{pmatrix}=\begin{pmatrix}-0.33\\0.34\\0.23\end{pmatrix}
\end{equation}
```

```python
print("2) Axe de rotation d'ordre 2 selon c")
print(" ")
S2 = SymmOp.from_xyz_string("-x,-x+y,-z")
print("Matrice de rotation associée :")
print(S2.rotation_matrix)
print("Vecteur de translation associé :")
print(S2.translation_vector)
print("Application de l'axe de rotation C2 à l'atome de Se (#3) situé en (x,y,z) = (0.67,0.33,0.77) :")
print(S2.operate((0.67,0.33,0.77)))
Image("C2.jpg")
```

```python
%%latex
Equation matricielle

\begin{equation}
\begin{pmatrix}-1 & 0 & 0\\-1 & 1 & 0\\0 & 0 & -1\end{pmatrix}\begin{pmatrix}x\\y\\z\end{pmatrix}=\begin{pmatrix}-x\\-x+y\\-z\end{pmatrix}
\end{equation}

Ce qui donne avec $(x,y,z) = (0.67,0.33,0.77)$ :

\begin{equation}
\begin{pmatrix}-1 & 0 & 0\\-1 & 1 & 0\\0 & 0 & -1\end{pmatrix}\begin{pmatrix}x\\y\\z\end{pmatrix}=\begin{pmatrix}-0.67\\-0.34\\-0.77\end{pmatrix}
\end{equation}
```

```python
print("3) Plan miroir (m) --> élément #6 du fichier CIF")
print(" ")
S3 = SymmOp.from_xyz_string("-y,-x,z")
print("Matrice de rotation associée :")
print(S3.rotation_matrix)
print("Vecteur de translation associé :")
print(S3.translation_vector)
print("Application du plan miroir m à l'atome de Sn (#6) situé en (x,y,z) = (0,1,1) :")
print(S3.operate((0,1,1)))
Image("mirror.jpg")
```

```python
%%latex
Equation matricielle

\begin{equation}
\begin{pmatrix}0 & -1 & 0\\-1 & 0 & 0\\0 & 0 & 1\end{pmatrix}\begin{pmatrix}x\\y\\z\end{pmatrix}=\begin{pmatrix}-y\\-x\\z\end{pmatrix}
\end{equation}

Ce qui donne avec $(x,y,z) = (0,1,1)$ :

\begin{equation}
\begin{pmatrix}0 & -1 & 0\\-1 & 0 & 0\\0 & 0 & 1\end{pmatrix}\begin{pmatrix}0\\1\\1\end{pmatrix}=\begin{pmatrix}-1\\0\\1\end{pmatrix}
\end{equation}
```
